```python
import pandas as pd
```

```python
df_demo = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/lessons/refs/heads/main/5_6_eda_inf_stats_tableau/project/files_for_project/df_final_demo.txt")

df_web_pt1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/lessons/refs/heads/main/5_6_eda_inf_stats_tableau/project/files_for_project/df_final_web_data_pt_1.txt")
df_web_pt2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/lessons/refs/heads/main/5_6_eda_inf_stats_tableau/project/files_for_project/df_final_web_data_pt_2.txt")

df_web = pd.concat([df_web_pt1, df_web_pt2])

df_variation = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/lessons/refs/heads/main/5_6_eda_inf_stats_tableau/project/files_for_project/df_final_experiment_clients.txt")
```

```python
numeric_cols_demo = ['client_id', 'clnt_tenure_yr', 'clnt_tenure_mnth', 'clnt_age',
       'num_accts', 'bal', 'calls_6_mnth', 'logons_6_mnth']
```

```python
for col in numeric_cols_demo:
    median_value = df_demo[col].median()
    df_demo[col].fillna(median_value, inplace=True)

gendr_mode = df_demo["gendr"].mode()[0]
df_demo["gendr"].fillna(gendr_mode, inplace=True)
```
